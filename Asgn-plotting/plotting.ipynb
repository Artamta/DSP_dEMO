```python
import pandas as pd

a = pd.read_csv("/Users/ayush/Desktop/dsp/DSP_dEMO/Asgn-plotting/iip_wrt_prevyr.csv") 
print(a.head)
```

```python
year = a.Year
mining = a.Mining
manf = a.Manufacturing
ele = a.Electricity
gen = a.General

fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

ax.axhline(y=0, color='black',linestyle="--")   # X-axis
#ax.axvline(x=x_center, color='black', linewidth=1.5)
ax.bar(year,mining,label = "Mining",color='LightSkyBlue')

ax.bar(year,manf,bottom=mining, label = "Manuf",color='IndianRed')

ax.bar(year,ele,bottom=mining+manf,label = "Electricity",color='LightSalmon')

ax.bar(year,gen,bottom=mining+manf+ele,label = "Genral",color='Moccasin')

ax.legend(ncols=2)

ax.set_title("IP GROWTH FOR INDIA", fontsize=14)
ax.set_xlabel("Finential Year")
ax.set_ylabel("IP Gwowth Rate wrt Prev Year",fontsize=12,loc='bottom')
```
